# KNN Algorithm on Breast Cancer Dataset

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
```

```python
#Read the .csv file
df=pd.read_csv('BrCancer.csv')
df.head()
```

```python
#Using describe() and info() methods
df.info()
```

```python
df.describe()
```

```python
#Removing unnecessary columns
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
#Checking for the Diagonisis column
print(df['diagnosis'].value_counts())
plt.pie(df['diagnosis'].value_counts(),labels=['Benign','Malignant'],autopct='%2.2f%%')
plt.show()
```

```python
#Splitting the dataset
print(df.columns)
X=df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean', 'symmetry_mean', 'fractal_dimension_mean','radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se','compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se','fractal_dimension_se', 'radius_worst', 'texture_worst','perimeter_worst', 'area_worst', 'smoothness_worst','compactness_worst', 'concavity_worst', 'concave points_worst','symmetry_worst', 'fractal_dimension_worst']]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
#Using KNN algorithm on the dataset
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)   #Here we take n as 2 because we want 2 clusters
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
#Measuring the accuracy
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
